# Preproceso del foro laweb sobre temas de computación

Para construir el vocabulario necesario para construir el chabot se descargaron las páginas de los temas más relevantes del foro [la web del programador](www.lawebdelprogramador.com)

## Cargue de librerias y ubicación de páginas almacendas

```python
import os
import numpy as np
import pandas as pd
import requests 
from bs4 import BeautifulSoup 
from pathlib import Path
import json
```

Los archivos se descargaron todos en una carpeta:

```python
directory_path = 'wwwlawebdelprogramadorcom/foros/'
```

En el directorio se descargaron varios temas, por lo que quedaron a su vez, separados entre si, por lo que es necesario hacer un mapeo de todos los archivos descargados

```python
subdir_list = os.listdir(directory_path)
subdir_list 
```

## Recopilación de la información

```python
directorio = 'wwwlawebdelprogramadorcom/foros/'
subdirectory_path_3 = 'Algoritmia/'
preguntas = []
respuestas = []
for subdirectorio in subdir_list:
    subdirectory_path = Path(directorio+subdirectorio)
    lista_archivos = os.listdir(subdirectory_path)
    subdirectorio = subdirectorio+"/"
    for archivo in lista_archivos:
        ruta_archivo = Path(str(directorio + subdirectorio + archivo ))
        enlace = open(ruta_archivo)
        print(ruta_archivo)
        try:
            lectura =BeautifulSoup(enlace,'html.parser',from_encoding="utf-8")  
            resultados = lectura.find_all("div", class_="list_description")
            n = len(resultados)
            bloques =[]
            cont = 0
            if n>2:
                for block in resultados:
                    edited = block.get_text().replace('\n',"").replace('\t',"")
                    control = 'Última edición'
                    validador = (control in edited)
                    if validador == False :
                        if cont == 0:
                            pregunta = edited
                        else:  
                            bloques.append(edited)
                    cont +=1
                preguntas.append(pregunta)
                respuestas.append(bloques)
        except:
            pass
```

```python
print(preguntas[8],'\n\n')
print(preguntas[98])
```

```python
diccionario = dict(zip(preguntas, respuestas))
```
