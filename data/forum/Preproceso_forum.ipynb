## Cargue de librerias y ubicación de páginas almacendas

```python
import os
import numpy as np
import pandas as pd
import requests 
from bs4 import BeautifulSoup 
from pathlib import Path
import json
from collections import defaultdict
```

```python
directorio = 'forum/'
```

__Lectura de los archivos en la carpeta__

```python
archivos = os.listdir(directorio)
archivos
```

```python
len(archivos)
```

### Recopilación de la información

```python
tematicas = []
preguntas = []
respuestas = []
cont = 0
for archivo in archivos:
    ruta_archivo = Path(str(directorio + archivo ))
    enlace = open(ruta_archivo)
    lectura = BeautifulSoup(enlace,'html.parser',from_encoding="utf-8")
    pregunta = lectura.find_all("div", class_="for_topic__question typo_content")
    #print( "1" ,archivo, type(pregunta))
    pregunta = pregunta[0].get_text().replace('\n',"").replace('\t',"")

    #print("2", archivo, pregunta)
    try:
        #print("1", archivo, pregunta)
        res=[]
        tematica = None
        lista_tematica = lectura.find('div', class_="grid_line").get_text().split("\n")
        m = len(lista_tematica)
        if m > 0:
            res = [i for i in lista_tematica if i]
            n = len(res)
            tematica = res[n-1]
        else:
            tematica = str("Varios")

        #print("2", archivo, tematica)
        pregunta = lectura.find_all("div", class_="for_topic__question typo_content")
        pregunta = pregunta[0].get_text().replace('\n',"").replace('\t',"")
        resultados = lectura.find_all("div", class_="for_answer__content")
        n = len(resultados)

        if n>0:
            print("3", n, archivo)
            
            bloques =[]
            for i in range(n):
                
                edited = resultados[i].get_text().replace('\n',"").replace('\t',"")
                bloques.append(edited)
         #       print("4", archivo, edited)
                #print(res)
            tematicas.append(tematica)
            preguntas.append(pregunta)
            respuestas.append(bloques)
            #cont +=1            
            
        else:
            pass
            #print("3. No proceso:", ruta_archivo)

    except:
        pass
```

```python
diccionario = dict(zip(preguntas, respuestas))
```

```python
diccionario_anidado = defaultdict(dict)
for x, y, z in zip(tematicas, preguntas , respuestas):
    diccionario_anidado[x][y] = z
```
